```python
# realizamos el código de regex  r'\b[A-Z][.| ][ |.\ [A-Z]\S+'
Regex= r'\b[A-Z][.| ][ |.\ [A-Z]\S+'
```

```python
shark5['Investigator or Source']= shark5['Investigator or Source'].astype(str) #Pasamos el tipo a string.
```

```python
shark5['Source'].head(30) #comprobamos que regex ha funcionando.
```

```python
shark6=shark5.copy()
```

```python
def sharkre(string):
    result= re.findall(r'\b[A-Z][.| ][ |.\ [A-Z]\S+', string)
    try:
        return result[0]
    except:
        return None
#Realizamos esta función para conseguir separar la columna en el
#caso de los investigadores. Al dar un error por lista fuera de rango, usando el try y excep para que lo obvie y lo sustituya 
# por none
```

```python
shark6['Researchers'] = shark6['Investigator or Source'].apply(sharkre)
```
